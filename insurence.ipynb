```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
X = df_reorder.drop(columns=['charges','region']) 

# Create a DataFrame to hold VIF results
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

we have high VIF but hopfully it's not higher than 10 . but `bmi` it's still conserning . mabe after scalling it fixed

i think building a version of our data is a good idea . one without `bmi`

```python
df_without_bmi = df_reorder.drop(columns="bmi")
```

let's go first with the `DF` that dosen't contain `bmi`

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
region_encoded = encoder.fit_transform(df_without_bmi[['region']])
region_encoded_df = pd.DataFrame(region_encoded, columns=encoder.get_feature_names_out(['region']))

# Join back to the original df (drop old 'region' if you want)
df_encoded = pd.concat([df_without_bmi.drop(columns=['region']), region_encoded_df], axis=1)

print(df_encoded.head())
```

```python
colunms_name =['age', 'sex', 'children', 'parent', 'smoker', 
       'region_northeast', 'region_northwest', 'region_southeast',
       'region_southwest','charges']
df_reorder = df_encoded.loc[:,colunms_name]
df_reorder
```

### data spliting and standardization

```python
from sklearn.model_selection import train_test_split
#some how there was a nan data , we cover it here
df_reorder = df_reorder.dropna()
x = df_reorder.drop(columns=['charges'])
y = df_reorder['charges']
x_train , x_test , y_train , y_test = train_test_split(x , y , random_state=42 , test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

## Model building
in this part we build 3 models (`RandomForestRegressor` , `XGBRegressor` , `KNeighborsRegressor`) and evaluate them at the end

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection  import GridSearchCV
reg = RandomForestRegressor()
param_dist = {
    'n_estimators': [100, 200, 500, 700 ,1000],
    'max_depth': [5, 10, 15, 20 , 25],
    'min_samples_split': [2, 5, 10 , 15],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
}
forest_search = GridSearchCV(estimator=reg,param_grid=param_dist , cv=5 , n_jobs=-1 , verbose=2)
forest_search.fit(x_train,y_train)
```

```python
forest_search.best_params_ , forest_search.best_score_
```

### XGBRegressor

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(objective='reg:squarederror', random_state=42)

# Define search space
param_XGB = {
    'n_estimators': [100, 300, 500, 700],
    'max_depth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.6, 0.7, 0.8],
    'colsample_bytree': [0.6, 0.7, 0.8],
    'gamma': [0, 0.1, 0.2, 0.5],
}
xgb_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_XGB,
    cv=5,                  
    verbose=2,
    n_jobs=-1              # Use all CPU cores
)
xgb_search.fit(x_train , y_train)
```

```python
xgb_search.best_params_ , xgb_search.best_score_
```

### KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
param_knn = {
    'n_neighbors': [3, 5, 7, 9, 11, 15, 20],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size': [10, 20, 30, 40],
    'metric': ['minkowski', 'euclidean', 'manhattan']
}
knn_search = GridSearchCV(
    estimator=knn,
    param_grid=param_knn,
    cv=5,                  
    verbose=2,
    n_jobs=-1              # Use all CPU cores
)
knn_search.fit(x_train , y_train)
```

```python
knn_search.best_score_
```

## Evaluation
evaluate our models with `R2_score` , `Mean_absolute_error` and `Mean_squared_error` over our test data

```python
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
from sklearn.model_selection import cross_val_score
models = {"randomforest" : forest_search,
          "XGB" : xgb_search,
          "knneighbors" : knn_search }
mae = []
mse = []
r2 = []
for model in models.values():
    y_pred = model.predict(x_test)
    mae.append(round(mean_absolute_error(y_test , y_pred) , 2))
    mse.append(round(mean_squared_error(y_test , y_pred),2))
    r2.append(str(round(r2_score(y_test , y_pred),2)*100)+"%")
```

```python
test_score = pd.DataFrame({"mean absolute error": mae ,
                            "mean squared error" : mse ,
                            "r2" : r2    
                           } , index=["randomforest" , "XGB" , "knneighbors"])
```

```python
test_score
```
